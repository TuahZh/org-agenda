## event horizon radius equal to the Schwarzschild radius
## r_s = 2GM/c^2
## density = 3M/(4*pi*r_s^3)
## density = 3/(32*pi) * (c^6/G^3) * M^-2

```python
c = 1.*2.e10 # cm/s
pi = 1.*3.1415
G = 1.*6.67e-8 # cm^3/g/s^2
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = np.arange(1000., 1000.*1000., 0.4)
dens = 3./(32.*np.pi)*(c**6/G**3)*m**(-2)
```

```python
plt.plot(m, dens)
plt.xscale('log')
plt.yscale('log')
```
